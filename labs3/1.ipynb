```python
import pandas as pd

# Загрузить данные из файла billing.csv.
df = pd.read_csv('billing.csv', sep=';',parse_dates=['Дата (коммутатор)', 'Время по Москве'], skipinitialspace = True,
dtype= {
'IMSI': 'object',
'IMEI': 'object', 
'Продолжительность': 'float', 
'Адрес первой БС': 'string',
'Номер': 'string', 
'Контакт': 'string', 
'A номер': 'string', 
'B номер': 'string', 
'Набрано': 'string', 
'ID первой БС': 'string',
'Переадресовано': 'string',
'LAC': 'object',
'SMS центр':'object'
})

print(df.info())
```

```python
# Выделить все сеансы связи, которые осуществлял абонент 79022810325 в период с 10.09.13 до 10.10.13 (включительно).
phone = "79022810325"
filter1 = '`Номер`== @phone or `Контакт`== @phone or `Переадресовано`== @phone or `B номер`== @phone or `A номер`== @phone'
filter2 = '`Время по Москве` >= datetime(2013, 9, 13) and `Время по Москве` <= datetime(2013, 10, 13)'
df.query(filter1 and filter2) # ответ
```

```python
# Определить распределение сеансов связи по дням недели для абонента 70887390016.
phone = "70887390016"
filter1 = '`Номер`== @phone or `Контакт`== @phone or `Переадресовано`== @phone or `B номер`== @phone or `A номер`== @phone'
df.query(filter1).groupby(df['Время по Москве'].dt.strftime('%A')).count()['A номер'] #ответ
```

```python
# Определить идентификаторы базовых станций, где были зафиксированы сеансы связи абонента 79217570540.
phone = "79217570540"
filter1 = '`Номер`== @phone or `Контакт`== @phone or `Переадресовано`== @phone or `B номер`== @phone or `A номер`== @phone'
df.query(filter1)['ID первой БС'].dropna().unique() #ответ
```

```python
# Определить распределение сеансов связи по часам в течении суток для абонента 79216666377.
phone = "79216666377"
filter1 = '`Номер`== @phone or `Контакт`== @phone or `Переадресовано`== @phone or `B номер`== @phone or `A номер`== @phone'
df.query(filter1).groupby(df['Время по Москве'].dt.strftime('%H')).count()['A номер'] #ответ
```
